```python
from datasets import load_dataset
dataset = load_dataset('THUDM/LongBench-v2', split='train')
```

```python
import random

# Filter contexts under 400k characters
short_contexts = dataset.filter(lambda x: len(x['context']) < 400000)

# Get a random example
random_example_context = random.choice(short_contexts)['context']

# Print the context
print(len(random_example_context))

print(random_example_context)
```

```python
from tqdm import tqdm

search_string = "Simulated great grief"
for idx, item in tqdm(enumerate(dataset)):
    if search_string in item['context']:
        print(f"Found at index {idx}")
        print("Context:", item['context'])
        print("-" * 80)
```

```python
def insert_text_every_n_newlines(text, insert_text, n=100):
    # Split the text into lines
    lines = text.split('\n')
    
    # Initialize result list and newline counter
    result = []
    newline_count = 0
    
    # Process each line
    for line in lines:
        result.append(line)
        newline_count += 1
        
        # Insert the text after every n newlines
        if newline_count % n == 0:
            result.append('')  # First newline
            result.append(insert_text)  # The text to insert
            result.append('')  # Second newline
    
    # Join the lines back together
    return '\n'.join(result)

# Example usage:
insert_text = 'Remember: Your task is to repeat the 5 sentences after "Van Aldin stared at him with awakening interest"'

# Assuming your text is in a variable called 'text'
modified_text = insert_text_every_n_newlines(random_example_context, insert_text)
```
